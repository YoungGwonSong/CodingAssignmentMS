Comparing the sample distribution with the theorectical distribution.

```python
import random
import matplotlib.pyplot as plt
import math
from scipy.stats import norm
from scipy.stats import chi2
import numpy as np
import csv
```

```python
with open('uniform.csv', 'r') as f:
  reader = csv.reader(f)
  s_uniform = list(reader)
s_uniform = s_uniform[1:]
s_uniform = [float(i[0]) for i in s_uniform]
#s_uniform[:10]
```

```python
print("length", len(s_uniform))
print("mean" , np.mean(s_uniform))

num_bins = 100
n, bins, patches = plt.hist(s_uniform, num_bins,
                            density = 1,
                            color ='green',
                             alpha = 0.7)

y = [1 for x in bins]

plt.plot(bins, y, '--', color ='black')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.title('Uniform Sample \n\n',
          fontweight = "bold")

plt.show()
```

```python
with open('normal.csv', 'r') as f:
  reader = csv.reader(f)
  s_normal = list(reader)
s_normal = s_normal[1:]
s_normal = [float(i[0]) for i in s_normal]
# s_normal[:10]
```

```python
print("length", len(s_normal))
print("mean", np.mean(s_normal))

num_bins = 100

n, bins, patches = plt.hist(s_normal, num_bins,
                            density = 1,
                            color ='green',
                            alpha = 0.7)

y = ((1 / (np.sqrt(2 * np.pi) )) *
     np.exp(-0.5 * (bins )**2))

plt.plot(bins, y, '--', color ='black')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.title('Normal distribution \n\n',
          fontweight = "bold")

plt.show()
```

```python
with open('chi.csv', 'r') as f:
  reader = csv.reader(f)
  s_chi = list(reader)
s_chi = s_chi[1:]
s_chi = [float(i[0]) for i in s_chi]
# s_chi[:10]
```

```python
print("length", len(s_chi))
print("mean" , np.mean(s_chi))
num_bins = 100
n, bins, patches = plt.hist(s_chi, num_bins,
                            density = 1,
                            color ='green',
                            alpha = 0.7)

y = [chi2.pdf(x, 2) for x in bins]

plt.plot(bins, y, '--', color ='black')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.title('Chi-Squared df2 Sample \n\n',
          fontweight = "bold")

plt.show()
```
